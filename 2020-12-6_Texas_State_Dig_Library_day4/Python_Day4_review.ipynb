## Data Aggregation using `pandas`

1. Use `describe()` for summary statistics
2. Dealing with missing values
3. Use `groupby()` to explore data further

```python
# 1 - summary stats
df.describe()
```

```python
# 2 - finding missing values
df.isnull().sum()
```

```python
df.fillna(0).isnull().sum()
```

```python
df['E19_period_use'].fillna(0).describe()
```

```python
# df = df['E19_period_use'].fillna(0)
df['E19_period_use'].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
# 3 - groupby()
pd.DataFrame(df.groupby(['C01_respondent_roof_type', 
            'C02_respondent_wall_type']).describe()['A11_years_farm']['count'])
```
