# 8 - Looping over darta sets -- automating tasks

```python
import glob
```

```python
print(glob.glob('gapminder*.csv'))
```

```python
for file in glob.glob('gapminder*.csv'):
    print(file)
```

```python
with open('loop_mintomax.txt','w') as output:
    output.write('{}\t{}\t{}\t{}\n'.format('filename','minimun','maximum','mean'))
    for file in glob.glob('gapminder*.csv'):
        dataframe=pd.read_csv(file,index_col='country')
        subset=dataframe.loc[:,'gdpPercap_1952']
        output.write('{}\t{}\t{}\t{}\n'.format(file.strip('.csv'),subset.min(),subset.max(),subset.mean()))
```
